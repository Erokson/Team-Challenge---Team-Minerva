```python
import pandas as pd
import seaborn as sns
from toolbox_ML import (describe_df, tipifica_variables,
                        get_features_num_regression, plot_features_num_regression)

# Cargar ejemplo de dataset 
df_titanic = sns.load_dataset('titanic')
# 1) Describir el dataset
desc = describe_df(df_titanic)
print(desc)

# 2) Tipificar las variables
tipos_sugeridos = tipifica_variables(df_titanic, umbral_categoria=10, umbral_continua=0.2)
print(tipos_sugeridos)

# 3) Seleccionar features numéricas para regresión (usamos 'fare' como target simulado)
features_num = get_features_num_regression(df_titanic, 'survived', umbral_corr=0.2, pvalue=0.05,tipo_variables=tipos_sugeridos)
print("Features numéricas candidatas:", features_num)

# 4) Pairplots de las features seleccionadas
plot_features = plot_features_num_regression(df_titanic, target_col='survived', columns=features_num, 
                                             umbral_corr=0.2, pvalue=0.05)
print("Features finalmente graficadas:", plot_features)




```
